Build an Artificial Neural Network by implementing the Backpropagation algorithm and test the same using appropriate data sets. Vary the ativation functions used and compare the results.

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python

x, y = datasets.load_iris(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40)

model = Sequential()
model.add(Dense(2, input_shape=(4,)))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.compile(loss='mean_squared_error', metrics=['accuracy'])
model.summary()
model.fit(x_train, y_train, batch_size=32, epochs=3)
score = model.evaluate(x_test, y_test)
print(score)
```

For Tanh Function:

```python

model = Sequential()
model.add(Dense(2, input_shape=(4,)))
model.add(Dense(1))
model.add(Activation('tanh'))
model.compile(loss='mean_squared_error', metrics=['accuracy'])
model.summary()
model.fit(x_train, y_train, batch_size=32, epochs=10)
score = model.evaluate(x_test, y_test)
print(score)
```

For ReLU Function:

```python

model = Sequential()
model.add(Dense(2, input_shape=(4,)))
model.add(Dense(1))
model.add(Activation('relu'))
model.compile(loss='mean_squared_error', metrics=['accuracy'])
model.summary()
model.fit(x_train, y_train, batch_size=32, epochs=10)
score = model.evaluate(x_test, y_test)
print(score)
```
