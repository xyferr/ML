```python
mat1 = np.random.randint(1,10,size=(3,3))
mat2 = np.random.randint(1,10,size=(3,3))

print("matrix 1:\n", mat1)
print()
print("matrix 2:\n", mat2)
print()
```

```python
print(mat1 @ mat2)
print()
print(np.dot(mat1, mat2))
print()
print(mat1.dot(mat2))
print()
print("element-wise multiplication:\n", mat1 * mat2)
print()
det = np.linalg.det(mat1)
print(f"determinant of matrix 1: {det:.2f}")
print()
eigvals, eigvecs = np.linalg.eig(mat1)
print("eigenvalues of matrix 1:", eigvals)
print()
print("eigenvectors of matrix 1:\n", eigvecs)
```

```python
arr6 = np.random.rand(5,4)

print(arr6,"\n")
```

```python
print(f"Mean: {np.mean(arr6):.2f} , Median: {np.median(arr6):.2f} , std: {np.std(arr6):.2f} \n")

print("Max per column:", np.max(arr6, axis=0))
print("Min per row:", np.min(arr6, axis=1))
print()
print(np.max(arr6))
print(np.min(arr6))
print(arr6.min(axis=0))

print()
# 3. Normalize (Min-Max Scaling)
# Formula: (x - min) / (max - min)
norm_arr6 = (arr6 - arr6.min()) / (arr6.max() - arr6.min())
print("Normalized array:\n", norm_arr6)

```

```python
arr7 = np.random.randint(1,100, size=(10))
print("Original array:", arr7)
```

```python
#sort
print("Sorted array:", np.sort(arr7))

# Indices of sorted array
sorted_indices = np.argsort(arr7)
print("Indices of sorted array:", sorted_indices)

#search
min_id = np.argmin(arr7)
print("Index of minimum value:", min_id)

#search using where
gt_50_indices = np.where(arr7 > 50)
print("Indices of elements greater than 50:", gt_50_indices[0])
max_id = np.where(arr7 == np.max(arr7))
print("Index of maximum value:", max_id)
max_id = np.argmax(arr7)
print("Index of maximum value using argmax:", max_id)
```

```python
arr8 = np.arange(1,37).reshape(6,6)
print("Original 6x6 array:\n", arr8)
print()
#alternate rows
print(arr8[::2,:],"\n")

#alternate columns
print(arr8[:,::2],"\n")

#main diagonal
print("Main diagonal:", np.diag(arr8),"\n")
#secondary diagonal
print("Secondary diagonal:", np.diag(np.fliplr(arr8)), "\n")

#diagonal without function
print("Diagonal: " , arr8[np.arange(6), np.arange(6)])

```

```python
base = np.random.randint(1, 10, (4, 4))
print("Original array:\n", base)
print()

# 1. Add row of zeros
row_zeros = np.zeros((1, 4)) # Shape must match columns
base = np.vstack([base, row_zeros])

print("After adding row of zeros:\n", base)
print()

# 2. Add column of ones
col_ones = np.ones((5, 1)) # Shape must match new row count (5)
base = np.hstack([base, col_ones])
print("After adding column of ones:\n", base)
print()

# 3. Delete 2nd row (index 1) and 3rd column (index 2)
# Note: Axis 0 is rows, Axis 1 is cols
base = np.delete(base, 1, axis=0) # Delete row
base = np.delete(base, 2, axis=1) # Delete col
print("Final Shape:", base.shape)
print("Array after deletions:\n", base)
```

```python
# Coefficients Matrix A
A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
])

# Constants Vector B
B = np.array([8, -11, -3])

# Solve
solution = np.linalg.solve(A, B)
print(f"Solution: x={solution[0]:.2f}, y={solution[1]:.2f}, z={solution[2]:.2f}")

# Verify: A dot X should equal B
print("Check (Ax - B):", np.allclose(np.dot(A, solution), B))
print(np.dot(A, solution), "==" , B)
```
