```python
import numpy as np
```

```python
arr1 = np.random.randint(1,51, size=(5,5))
print(arr1)
print()

```

```python
print("second row", arr1[1,:])
print()
print("third column", arr1[:,2])
print()
print(arr1[:3,:3])
print()
print(arr1[arr1>30])
```

```python
arr2 = np.arange(1,21)
print(arr2)
```

```python
even_num = arr2[arr2%2==0]
print("even numbers:", even_num)
print()
arr2[arr2%3==0]= -1
print(arr2)
print()
arr2_new = np.arange(1,21)
gtt_arr2 = arr2_new[arr2_new>10]
print("greater than 10:", gtt_arr2)
print()
```
